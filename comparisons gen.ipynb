```python
Frames_Path1 = '/content/drive/MyDrive/Daming Colab/Ranking/Frames'
Frames_Path2 = '/content/drive/MyDrive/Daming Colab/Ranking/FramesNew'

Frames_Path3 = '/content/drive/MyDrive/Daming Colab/Ranking/Frames_diff_background'
Frames_Path4 = '/content/drive/MyDrive/Daming Colab/Ranking/Frames_diff_background1'
Frames_Path5 = '/content/drive/MyDrive/Daming Colab/Ranking/Frames_diff_background2'

Frames_Path6 = '/content/drive/MyDrive/Daming Colab/Ranking/Frames_pandaside'
Frames_Path7 = '/content/drive/MyDrive/Daming Colab/Ranking/Frames_lyingdown'

dataset_path1 = '/content/drive/MyDrive/Daming Colab/Ranking/mix_train_data_set_30000.csv'
dataset_path2 = '/content/drive/MyDrive/Daming Colab/Ranking/mix_train_data_set_16000.csv'
dataset_path3 = '/content/drive/MyDrive/Daming Colab/Ranking/mix_train_data_set_24000.csv'
```

```python
def comparisons_gen(Frames_Path,dataset_path,comparisons_num):

    for files in  os.listdir(Frames_Path):
        Frames = os.listdir(Frames_Path)
        Frames_num=np.size(Frames)

    for i in tqdm(range(comparisons_num)):

        index=random.sample(range(1,Frames_num),2)

        if index[0]>index[1]:
                Label=1
        else:
                Label=0

        Frame_1=Frames[index[0]]
        Frame_2=Frames[index[1]]

        Data_Pairs=[Frames_Path+'/'+Frame_1 , Frames_Path+'/'+Frame_2 , Label]

        with open(dataset_path , 'a' , newline='') as f: 

          row=[Data_Pairs[0],Data_Pairs[1],Data_Pairs[2]]
          write=csv.writer(f)
          write.writerow(row)
```

```python
# 13000 for normal simple videos
# 6500 each
comparisons_gen(Frames_Path1,dataset_path1,6500)
comparisons_gen(Frames_Path2,dataset_path1,6500)
# 12000 for normal videos with backgrounds
# 4000 each
comparisons_gen(Frames_Path3,dataset_path1,4000)
comparisons_gen(Frames_Path4,dataset_path1,4000)
comparisons_gen(Frames_Path5,dataset_path1,4000)
# 5000 for normal videos with panda side or lyingdown
# 2500 each
comparisons_gen(Frames_Path6,dataset_path1,2500)
comparisons_gen(Frames_Path7,dataset_path1,2500)
```
