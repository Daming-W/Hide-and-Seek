```python
 # -*- coding: utf-8 -*-
"""
Created on Thu Apr 22 00:59:43 2021

@author: yoo
"""

import cv2
import os

from google.colab import drive
drive.mount('/content/drive')

def video2frames(videoPath, imgPath):
    if not os.path.exists(imgPath):
        os.makedirs(imgPath)             # if the file does not exist then create a ner one
    cap = cv2.VideoCapture(videoPath)    # read the video
    success = cap.isOpened()         
    print(success)
    fps = cap.get(cv2.CAP_PROP_FPS)      # read fps
    print('fps:',fps)

    frames = 1                           # count numbers
    count = 1                            

    while(success):
        flag, frame = cap.read()         # read each single frames
        if not flag:
            print(flag)
            print("Process finished!")
            break
        else:
            if frames % 1 == 0: # take 1 frame in 1 frame
                frame = cv2.resize(frame,(224,224))
                imgname = 'frame' + str(count).rjust(3,'0') + ".jpg"
                newPath = imgPath + imgname
                print(imgname)
                cv2.imwrite(newPath, frame, [cv2.IMWRITE_JPEG_QUALITY, 100])
                # cv2.imencode('.jpg', frame)[1].tofile(newPath)
                count += 1
        frames += 1
    cap.release()
    print("there are %d frames" %(count-1))
          

```
